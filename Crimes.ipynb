```python
import pandas as pd
```

```python
crimedata = pd.read_csv("crimefile.csv")
```

```python
meanhouses = pd.read_csv("meanhouses.csv")
meanhouses
```

```python
crimedata.head(40)
```

```python
cols2016 = ['201601',
 '201602',
 '201603',
 '201604',
 '201605',
 '201606',
 '201607',
 '201608',
 '201609',
 '201610',
 '201611',
 '201612']
```

```python
crimedata["2016"]=crimedata[cols2016].sum(axis=1)
```

```python
crimedata.head(5)
```

```python
groupedcrime = crimedata.groupby(["Major Category","Borough"]).agg(sum)
```

```python
groupedcrime.head(15)
```

```python
groupedcrime.loc["Burglary"]
```

## Burglary 

```python
burglary_houseprice = pd.merge(left=groupedcrime.loc["Burglary"],right=meanhouses, how="inner", left_index=True, right_on="Area")
```

```python
burglary_houseprice = burglary_houseprice[1:]
```

```python
burglary_houseprice["2016_y"]=pd.to_numeric(burglary_houseprice["2016_y"])
```

```python
%matplotlib inline
burglary_houseprice.plot.scatter(x="2016_x",y="2016_y")
```

```python
burglary_houseprice["2016_x"].corr(burglary_houseprice["2016_y"])
```

## Crime Types

```python
import matplotlib.pyplot as plt 
n = 0
for crimeType in ["Burglary","Drugs","Robbery", "Criminal Damage", "Sexual Offences", "Violence Against The Person", "Fraud & Forgery","Theft & Handling"]:
    n += 1
    plt.subplot(2, 7, n)
    print(crimeType)
    crimeType_houseprice = pd.merge(left=groupedcrime.loc[crimeType],right=meanhouses, how="inner", left_index=True, right_on="Area")
    crimeType_houseprice = crimeType_houseprice[1:]
    crimeType_houseprice["2016_y"]=pd.to_numeric(crimeType_houseprice["2016_y"])
    %matplotlib inline
    crimeType_houseprice.plot.scatter(x="2016_x",y="2016_y")
    plt.ylabel('Average House Price')
    plt.xlabel("Number of Crimes Committed")
    plt.show()
```
